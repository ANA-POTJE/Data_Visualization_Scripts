```python
# Path of the file to read
ign_filepath = "data/ign_scores.csv"
ign_data = pd.read_csv(ign_filepath,index_col="Platform")

ign_data#.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8, 6))
# Bar chart showing average score for racing games by platform
sns.barplot(x=ign_data['Racing'], y=ign_data.index)
# Add label for horizontal axis
plt.xlabel("")
# Add label for vertical axis
plt.title("Average Score for Racing Games, by Platform")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,10))
# Heatmap showing average game score by platform and genre
sns.heatmap(ign_data, annot=True)
# Add label for horizontal axis
plt.xlabel("Genre")
# Add label for vertical axis
plt.title("Average Game Score, by Platform and Genre")
```

```python
##################
#### exercise DISTRIBUTION

cancer_b_filepath = "data/cancer_b.csv"
cancer_m_filepath = "data/cancer_m.csv"

cancer_b_data = pd.read_csv(cancer_b_filepath,index_col="Id")
cancer_m_data = pd.read_csv(cancer_m_filepath,index_col="Id")

cancer_b_data.head()
```

```python
cancer_m_data.head()
```

```python
sns.distplot(a=cancer_m_data['Area (mean)'], label='Malignant', kde=False)
sns.distplot(a=cancer_b_data['Area (mean)'], label='Belign', kde=False)
plt.title("Area (mean) for both benign and malignant cancer")
plt.legend()
```

```python
sns.kdeplot(data=cancer_m_data['Radius (worst)'], label='Malignant', shade=True)
sns.kdeplot(data=cancer_b_data['Radius (worst)'], label='Benign', shade=True)
plt.title("Radius (worst) for both benign and malignant cancer")
```
